# Data Visualisation

```python
import warnings
warnings.filterwarnings('ignore')
```

### Importing Packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('heart.csv')
```

### Let's See Heatmap of the dataset

```python
# Heatmap plot diagram
fig, ax = plt.subplots(figsize=(15,10))
sns.heatmap(data.corr(), ax=ax, annot=True)
print('These parameters are more related ')
```

### Pair Plot

```python
sns.pairplot(data)
plt.show()
```

### High Price Distribution.

```python
fig, ax = plt.subplots(figsize=(15,6))
sns.boxplot(data.chol, ax =ax)
plt.title("Cholesterol Distribution")
plt.show()
```

### Daily deviations of high price

```python
plt.scatter(data['chol'],data['thalachh'])
plt.xlabel('Cholesterol')
plt.ylabel('Maximum heart rate achieved')
plt.title('Heart rate related to Cholesterol')
plt.show()
```

### This is the relation between Age and resting blood pressure

```python
plt.scatter(data['age'],data['trtbps'])
plt.show()
```

### Relation between myocardial Infaraction with relation to Age.

```python
plt.scatter(data['age'],data['output'])
plt.show()
```
